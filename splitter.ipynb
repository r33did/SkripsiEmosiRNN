```python
import pandas as pd
import os
from scipy import signal
import matplotlib.pyplot as plt
```

```python
root = 'IAPS hari 1'
nama_klien = str(input("Nama klien = "))
nama_file = nama_klien+'.csv'
dirs = os.listdir(root)
print(nama_klien)
print(nama_file)
```

```python
pipi = []
alis = []
temp1 = []
temp2 = []
count = 0
df = pd.read_csv(root+'/'+nama_file)
pipi = list(df.iloc[:,2])
alis = list(df.iloc[:,3])
for i in range(0,31):
    print('---------------------------------------------------------')
    print(i)
    temp1.clear()
    temp2.clear()
    for j in range((count*1000),(1000*(count+1))):
        temp1.append(pipi[j])
        temp2.append(alis[j])
    nama_file = nama_klien+str(i)+'.csv'
    finaldirs = os.path.join(root,nama_file)
    df1 = pd.DataFrame({'Pipi':temp1,'Alis':temp2})
    df1.to_csv(finaldirs,mode='w+')
    count += 1
```

```python
import pandas as pd
import os
from scipy import signal
import matplotlib.pyplot as plt
data1n = []
data2n = []
root = 'Filtered'
emosi = ['NH','NL','PH','PL']
# pasien = ['adit','agus','amin','eka','hasna','riznop']
pasien = ['adit','agus','amin','bagus','basith','eka','hanif','rizki']
    
def lowpass_filter(sinyal,fcl):
    sampleRate = 200
    wnl = fcl/(sampleRate)
    b,a = signal.butter(3,wnl,'lowpass')
    fil = signal.filtfilt(b, a, sinyal)
    return fil

def filtering(folder):
    print("Filter dimulai, harap tunggu sebentar")
    dirs = os.listdir(folder)
    for y in pasien:
        for j in emosi:
            for z in range(1,6):
                df = pd.read_csv(folder+'/'+y+j+str(z)+'.csv')
                print(y+j+str(z))
#                 wk = df["Waktu"]
                pp = df['Pipi']
                al = df['Alis']
#                 wkt = list(wk)
                data1 = list(pp)
                data2 = list(al)

                t = [i for i in range(len(data1))]
                w = lowpass_filter(data1,2.0)
                x = lowpass_filter(data2,2.0)

                mn1 = min(w)
                mx1 = max(w)
                mn2 = min(x)
                mx2 = max(x)

                for i in range(len(w)):
                    data1n.append(abs((w[i]-mn1)/(mx1-mn1)))
                    data2n.append(abs((x[i]-mn1)/(mx1-mn1)))

                f = plt.figure()
                plt.xlabel('Data ke-')
                plt.ylabel('mV')
                plt.grid(True)
                plt.title(j+str(z))
                plt.plot(t,data1n)
                plt.plot(t,data2n)
                plt.savefig('Data_Plot/'+j+str(z)+'.png')
                f.clear()
                plt.close(f)

                d_t = list(zip(data1n,data2n))
                root_filter = 'IAPS2_filter'
                finaldirs = os.path.join(root_filter,y+j+str(z)+'.csv')
                df1 = pd.DataFrame(d_t,columns=['Pipi','Alis'])
                df1.to_csv(finaldirs)
                data1n.clear()
                data2n.clear()
    print('Filter Selesai !')
```

```python
filtering('IAPS hari 2')
```
